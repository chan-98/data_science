**Studying the relation between countrywise Covid19 infection and World Happiness Metrics.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Import Dataset

```python
dataset = pd.read_csv("/content/covid19_Confirmed_dataset.csv")
# confirmed covid cases on different dates from Jan to April
dataset.head(10)
```

```python
dataset.shape
```

```python
# Clean data

    # delete cols latitude and longitude

df1 = dataset.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
    # aggregate rows by country
covid_dataset_aggregated = dataset.groupby("Country/Region").sum()
covid_dataset_aggregated.head()
```

```python
covid_dataset_aggregated.shape

# 100 days (cols), 187 countries (rows)
```

```python
# visualize data related to countries
covid_dataset_aggregated.loc["China"]
```

```python
covid_dataset_aggregated.loc["China"].plot()
covid_dataset_aggregated.loc["India"].plot()
covid_dataset_aggregated.loc["Singapore"].plot()
plt.legend()
```

```python
# calculate a good measure (number) that can represent the spread of covid in a country
covid_dataset_aggregated.loc["China"].plot()
```

```python
covid_dataset_aggregated.loc["China"][:3].plot()  # spead in first 3 days
```

```python
# calculate the first derivative of the curve

covid_dataset_aggregated.loc["China"].diff()
```

```python
covid_dataset_aggregated.loc["China"].diff().plot()
```

The first derivative shows the rate of infection spread between Jan and April in China

```python
# max infection rate
covid_dataset_aggregated.loc["China"].diff().max()
```

```python
covid_dataset_aggregated.loc["India"].diff().max()
```

```python
covid_dataset_aggregated.loc["Vietnam"].diff().max()
```

```python
# max infec rate for all countries

countries = list(covid_dataset_aggregated.index)    # countries are stored in index
max_infec_rates = [covid_dataset_aggregated.loc[country].diff().max() for country in countries]
covid_dataset_aggregated["Maximum Infection Rates"] = max_infec_rates
```

```python
covid_dataset_aggregated
```

```python
# Now we no longer need all the individual data

covid_data = pd.DataFrame(covid_dataset_aggregated["Maximum Infection Rates"])
covid_data
```

```python
# Import World happiness data

happiness_report = pd.read_csv("/content/worldwide_happiness_report.csv")
happiness_report
```

```python
# Clean Data

drop_columns = ["Overall rank", "Generosity", "Score", "Perceptions of corruption"]
happiness_report.drop(drop_columns, axis=1, inplace=True)
```

```python
happiness_report.head()
```

```python
# make country/region as index
happiness_report.set_index("Country or region", inplace=True)
happiness_report.head()
```

Join `covid_data` and `happiness_report` using INNER JOIN

```python
covid_data.shape
```

```python
happiness_report.shape
```

```python
data = covid_data.join(happiness_report, how = "inner")
```

```python
data
```

Create a Correlation Matrix

```python
# Pearson Correlation Coefficient

data.corr()
```

From above matrix, we can say all of these show POSITIVE CORRELATION.

Max Infec Rates has very little correlation with Freedom to make life choices. Healthy life expectancy has very high correlation with per-capita GDP.
