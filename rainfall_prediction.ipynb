# Rainfall Prediction Using Linear Regression

DATA:

> "/content/austin_weather.csv"

```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/austin_weather.csv")
data
```

Cleaning the dataset.

> Remove text fields, only numbers are required

```python
# cleaning

data = data.drop(["Events", "Date", "SeaLevelPressureLowInches"], axis=1)
```

```python
data = data.replace("T", 0.0)
data = data.replace("-", 0.0)
```

```python
data
```

```python
data.to_csv("autin_weather_final.csv")
```

```python
data = pd.read_csv("autin_weather_final.csv")
data.head()
```

The last column "PrecipitationSumInches" will be the label based on which the model will be trained.

```python
X = data.drop("PrecipitationSumInches", axis=1)
```

```python
X
```

```python
Y = data["PrecipitationSumInches"]
Y
```

Reshape into a 2D vector

```python
Y = Y.values.reshape(-1, 1)
Y
```

Pick any random day from X and plot graph to observe the selected day

```python
day_index = 798
days = [i for i in range(Y.size)]
days
```

```python
# Initialise the LR classifier

classifier = LinearRegression()

# train the classifier

classifier.fit(X, Y)    # pass labels to train the model
```

```python
# plot graph of precipitation vs total num of days

print("Precipitation Trend Graph: \n")
plt.scatter(days, Y, color = 'g')
plt.scatter(days[day_index], Y[day_index], color = 'red')
plt.title("Precipitation Levels")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
# filter out
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
x_vis
```

```python
print("Precipitation vs. Attributes Trend Graph: \n")

for i in range(x_vis.columns.size):
  plt.subplot(3,2,i+1)
  plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color = 'g')
  plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color = 'red')
  plt.title(x_vis.columns.values[i])

plt.show()
```

The rainfall can be expected when temp and humidity are high. Not the scale of X and Y axes are different for different plots.
